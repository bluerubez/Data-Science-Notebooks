###### This short notebook shows how to use group aggregation functions as well as cross tabulations and pivot tables. The dataset used is a record of collected tips by a waitress or waiter at a particularal restaurant.

```python
!head tips.csv
```

```python
import pandas as pd
import warnings;warnings.filterwarnings('ignore')
tips = pd.read_csv('tips.csv');tips.head()
```

###### Making a new column which is the tip percentage of the bill. Also figuring out how many observations are in the table

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
len(tips.index)
```

###### Creating a groupby object to calculate the mean tip of male and female smokers and non-smokers

```python
grouped = tips.groupby(['sex', 'smoker']);grouped
```

```python
grouped_pct = grouped['tip_pct']
```

```python
grouped_pct.agg('mean')
```

###### We can then add another layer on top of the grouped object by further grouping in the total bill and applying multiple functions like so

```python
functions = ['count', 'mean', 'max']
result = grouped['tip_pct', 'total_bill'].agg(functions);result
```

###### And Create our own functions. This one just sorts by column then shows the biggest tips

```python
def top(df, n=5, column='tip_pct'):return df.sort_index(by=column)[-n:]
```

```python
top(tips)
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=1,column='total_bill')
```

```python
tips.pivot_table(['tip_pct', 'size'], rows=['sex', 'day'],cols='smoker', margins=True)
```

```python
pd.crosstab([tips.time, tips.day], tips.smoker, margins=True)
```
